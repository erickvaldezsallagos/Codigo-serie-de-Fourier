#Serie de Fourier

Gráficar la función original (a trozos) en el intervalo dado.

Graficamos la función a trozos

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x1 = np.linspace(-2, 0, 500)
x2 = np.linspace(0, 2, 500)

# Definir la función a trozos
y1 = np.zeros_like(x1)  # y = 0 cuando -2 <= x < 0
y2 = 2 - x2  # y = 2 - x cuando 0 <= x <= 2

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(x1, y1, label=r"$f(x) = 0, \, -2 \leq x < 0$", color="blue")
plt.plot(x2, y2, label=r"$f(x) = 2 - x, \, 0 \leq x \leq 2$", color="red")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Gráfica de la función a trozos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función que representa la serie de Fourier truncada
def fourier_series(x, N):
    result = 1/2  # Primer término de la serie (1)
    for n in range(1, N + 1):
        term1 = (2 / (n**2 * np.pi**2)) * (1 - (-1)**n) * np.cos(n * np.pi * x / 2)
        term2 = (2 / (n * np.pi)) * np.sin(n * np.pi * x / 2)
        result += term1 + term2
    return result

# Parámetros
x_vals = np.linspace(-2, 2, 500)

# Graficar la función con varios números de armónicos
plt.figure(figsize=(10, 7))

# Usar un ciclo for para graficar la aproximación con diferentes números de armónicos
for N_terms in [1, 5, 10, 20, 50]:  # Diferentes números de armónicos
    y_vals = fourier_series(x_vals, N_terms)
    plt.plot(x_vals, y_vals, label=f"{N_terms} armónicos")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Aproximación de la serie de Fourier con varios armónicos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

Usamos varios armónicos para aproximar mejor a la SERIE

EN EL LIMITE POR LA IZQUIERDA CONVERGE A $y=0$ y $x=-2$

EN LA DISCONTINUIDAD CONVERGE AL PROMEDIO

EL LIMITE POR LA DERECHA CONVERGE A $x=2$ Y $y=0$

```python
import numpy as np
import matplotlib.pyplot as plt

# Función a trozos
def f(x):
    return np.piecewise(x, [x < 0, x >= 0], [0, lambda x: 2 - x])

# Función que representa la serie de Fourier truncada
def fourier_series(x, N):
    result = 1/2  # Primer término de la serie (1)
    for n in range(1, N + 1):
        term1 = (2 / (n**2 * np.pi**2)) * (1 - (-1)**n) * np.cos(n * np.pi * x / 2)
        term2 = (2 / (n * np.pi)) * np.sin(n * np.pi * x / 2)
        result += term1 + term2
    return result

# Parámetros
x_vals = np.linspace(-2, 2, 500)

# Graficar la función a trozos y la serie de Fourier
plt.figure(figsize=(10, 7))

# Graficar la función a trozos
y_vals_function = f(x_vals)
plt.plot(x_vals, y_vals_function, label="Función a trozos", color='blue', linewidth=2)

# Usar un ciclo for para graficar la aproximación con diferentes números de armónicos
for N_terms in [1, 5, 10, 20, 50]:  # Diferentes números de armónicos
    y_vals_fourier = fourier_series(x_vals, N_terms)
    plt.plot(x_vals, y_vals_fourier, label=f"Serie de Fourier con {N_terms} armónicos")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Función a trozos y su serie de Fourier truncada")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
