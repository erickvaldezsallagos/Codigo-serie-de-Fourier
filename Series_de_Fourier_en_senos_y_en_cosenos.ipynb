#EJERCICIO DE EJEMPLO

```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
```

```python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
```

```python
plot(f, (x,-2,2),line_color ='red')
```

```python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

#Primer Ejercicio

$ f(x)=e^{2x} $ con $ 0\leq x \leq 1 $ en cosenos

Código de la función

```python
from sympy import  E, log

f1 = Piecewise(
    #(0, x < 0),
    (E**(2*x), (x >= 0) & (x <= 1)),
    (E**(-2*x), (x >= -1) & (x < 0))
)
```

Código de la gráfica

```python
plot(f1, (x,-1,1),line_color ='blue')
```

Gráfica de $ f(x)=e^{2x} $ con $ 0\leq x \leq 1 $ en cosenos

```python
s = fourier_series(f1, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f1, s1, s2, s3, (x, -1, 1), show=False, legend=True)
```

Mostrar serie

```python
s
```

Código de gráfica de la serie de Fourier en cosenos con armónicos

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

Gráfica de serie de Fourier en cosenos

```python
p.show()
```

#Ejercicio 2

Código de la función

```python
from sympy import  E
f2 = Piecewise(
    (E**(2*x), (x > 0) & (x <= 1)),
    (-E**(-2*x), (x >= -1) & (x <= 0))
)
```

Código de la gráfica

```python
plot(f2, (x,-1,1),line_color ='blue')
```

Gráfica de la función $ f(x)=e^{2x} $ con $ 0 \leq x \leq 1 $ en senos

Código de la serie de Fourier en senos

```python
s = fourier_series(f2, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f2, s1, s2, s3, (x, -1, 1), show=False, legend=True)
```

Serie de Fourier en Senos

```python
s
```

```python
p.show()
```

Códifo de la función a trozos en cosenos

Código de la gráfica de la función a trozos en cosenos

```python
plot(g, (x,-pi,pi),line_color ='blue')
```

Código de la serie de Fourier en cosenos

```python
s = fourier_series(g, (x, -pi, pi))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(g, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```
