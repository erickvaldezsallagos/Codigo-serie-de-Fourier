```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x') #introducimos una variable
u = sp.Function('u') #esto es una funcion
```

```python
a, b = sp.symbols('a b')

#Funcion de prueba u_n(x) que cumple u(0)=u(1)=0

u_n = a*x*(1-x) + b*x**2*(1-x)
```

```python
E = sp.Function('E') # esto es una funcion
```

```python
u_n_2 = sp.diff(u_n, x, 2)

E = u_n_2 + u_n
```
